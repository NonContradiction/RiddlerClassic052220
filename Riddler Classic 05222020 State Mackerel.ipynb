```python
#We have to strike spaces separately for the nine states with two words
thestates = [x.replace(' ', '') for x in thestates]
#We want lists of the unique letters in each state
#also lower case everything, of course, and alphabetize them
stateletters = [sorted(list(set(x.lower()))) for x in thestates]
stateletters
```

```python
#ALL-IMPORTANT PROGRESS BAR FUNCTION CODE : )
import time, sys
from time import sleep
from IPython.display import clear_output
import operator 
from collections import defaultdict

import functools

@functools.singledispatch
def normer(value):
    return 0
    
@normer.register(int)
def int_normer(value):
    return float(value)

@normer.register(float)
def float_normer(value):
    return value 

#Progress bar function: 

def update_progress_mine(progress, bar_length = 80):
    
    progress = normer(progress)
    if progress < 0:
        progress = 0
    if progress >= 1:
        progress = 1
    block = int(round(bar_length * progress))

    clear_output(wait = True)
    text = "Progress: [{0}] {1:.1f}%".format( "#" * block + "-" * (bar_length - block), progress * 100)
    print(text)
```

```python
#To (we imagine) help out the computer, we eliminate duplicate letters and alphabetize the ones that remain
wordletters = [sorted(list(set(x.lower()))) for x in words]
wordletters
```

```python
#A couple of dictionaries to help with our mackerel hunting: 
stateindices = {k: v for k, v in enumerate(thestates)}
allwordindices = {k: v for k, v in enumerate(words)}
```

```python
%%time
#So this is where the magic happens

#We make a dictionary that will store mackerel-Ohio pairs as we find them
mackdict = dict()

#for every word
for i, word in enumerate(wordletters):
    print(word)
    #for every state name
    #we count the exclusions
    excluscount = 0
    for j, state in enumerate(stateletters):
        if not any(item in word for item in state):
            excluscount += 1
            #This is a variable we only wind up using if excluscount remains at 1
            theuniquestate = j
            #We want it to be excluded exactly once to be a mackerel
    #and if it is . . .
    if excluscount == 1:
        newmack = allwordindices[i]
        mackdict[newmack] = stateindices[theuniquestate]
    
    update_progress_mine(i / len(wordletters))
```

```python
#We take a peek: 
mackdict
```

```python
#So of all the mackerels, which is the longest?
#Let's use a list comprehension and some lookups: 

mylengths = [len(x) for x in mackdict.keys()]
print(list(mackdict.keys())[mylengths.index(max(mylengths))])
print(mackdict[list(mackdict.keys())[mylengths.index(max(mylengths))]])
```

```python
#But what if there's actually more than one??!
#This code takes longer to run: 
longestmackerels = [x for x in mackdict.keys() if len(x) == max(mylengths)]
longestmackerels
```

```python
#So we print them all out: 
for eachmack in longestmackerels:
    print(eachmack, ': ', mackdict[eachmack])
```

```python
#If we want to just see a given state's mackerels, we use a list of lists, 
monsterlist = []
for eachstate in sorted(set(mackdict.values())):
    statelist = []
    for word, state in mackdict.items():  # for name, age in dictionary.iteritems():  (for Python 2.x)
        if state == eachstate:
            statelist.append(word)
    monsterlist.append(statelist)
```

```python
#One more Q: which state has the longest mackerels on average?

theaverages=[]
for eachlist in monsterlist:
    thelengths = [len(x) for x in eachlist]
    theavg = sum(thelengths)/len(thelengths)
    theaverages.append(theavg)
print(theaverages)
print('\nThe largest average of any state\'s mackerels is: ', max(theaverages))
largestmackerel = sorted(list(set(mackdict.values())))[theaverages.index(max(theaverages))]
print('That state is: ', largestmackerel)

largestlist = []
for word, state in mackdict.items():  # for name, age in dictionary.iteritems():  (for Python 2.x)
    if state == largestmackerel:
        largestlist.append(word)
print('And they are: ', largestlist)
```
